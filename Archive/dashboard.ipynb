# State Economic Indicators Dashboard

This dashboard loads pre-built chart HTML files exported by `Visualizations.ipynb`.  
**No chart code lives here** — to change a visualization, edit it in `Visualizations.ipynb` and re-run the export cell at the bottom.

To add a new chart: export it as HTML in `Visualizations.ipynb`, then add its filename to the `CHARTS` list below.

```python
import os
from IPython.display import display, HTML, IFrame

CHARTS_DIR = 'charts'

# ======================================================================
#  CHARTS — edit this list to add, remove, or reorder charts.
#  Each entry is ("Section Title", "filename.html").
# ======================================================================

CHARTS = [
    ("Interactive Choropleth Map",              "choropleth.html"),
    ("Animated Bubble Chart",                   "bubble_chart.html"),
    ("Correlation Heatmap",                     "correlation_heatmap.html"),
    ("Linked Brushing Dashboard",               "linked_brushing.html"),
    ("Top vs Bottom States by GDP per Capita",  "animated_bars.html"),
]

# -- Render all charts on one scrollable page --
for title, filename in CHARTS:
    filepath = os.path.join(CHARTS_DIR, filename)
    if not os.path.exists(filepath):
        display(HTML(f'<h2>{title}</h2>'
                     f'<p style="color:red;">File not found: {filepath}<br>'
                     f'Run the export cell in Visualizations.ipynb first.</p>'))
        continue

    display(HTML(f'<h2 style="margin-top:40px; border-bottom:2px solid #ccc; '
                 f'padding-bottom:8px;">{title}</h2>'))
    display(IFrame(src=filepath, width='100%', height=700))

display(HTML('<hr><p style="color:#888; text-align:center;">End of dashboard. '
             'Charts loaded from <code>Dashboard/charts/</code> &mdash; '
             'edit visuals in <code>Visualizations.ipynb</code> and re-run the export cell to update.</p>'))
```
